```python
def plot_dict(score_dict, title, word_sequence):
    score_sequence = [score_dict[token] for token in word_sequence]
    import numpy as np
    import matplotlib.pyplot as plt

    # x = np.random.rand(100).reshape(10,10)
    x = np.array(score_sequence).reshape(30, 30)
    plt.imshow(x, cmap=plt.cm.hot, vmin=0, vmax=1)
    plt.title(title)
    plt.colorbar()
    plt.savefig(f"{title}.svg")
    
    plt.show()
```

```python
from functools import partial

def score_iter_callback_fn(score_list, word_sequence, abstract_tokens, tf_score_dict, idf_score_dict, smi_score_dict):
    print(len(score_list))
    
    abstract_word_dict = {
        token: 1 if token in abstract_tokens else 0
        for token in word_sequence
    }
    plot_dict(abstract_word_dict, "abstract", word_sequence[:900])
    print(abstract_tokens)
    
    simp_count, diver_count, coh_count = 1, 1, 1
    
    for name, word_dict in [
        ("TF", tf_score_dict),
        ("IDF", idf_score_dict),
        ("SMI", smi_score_dict),
    ] + score_list:
        normalize(word_dict)
        if name == "init":
            name = "CenSMI"
        
        if name == "simp":
            name = f"Simplification_{simp_count}"
            simp_count += 1
        
        elif name == "div":
            name = f"Diversity_{diver_count}"
            diver_count += 1
        
        elif name == "coh":
            name = f"Coherence_{coh_count}"
            coh_count += 1

        for word, score in sorted(word_dict.items(), key=lambda x: -x[1])[:20]:
            print(word, score)
            
        plot_dict(word_dict, name, word_sequence[:900])

def normalize(word_dict):
    max_val = max(word_dict.values())
    for key in word_dict:
        word_dict[key] /= max_val

for i, item in enumerate(paper_items):
    if i != 45: continue
#     if i != 33: continue
    
    selected_paper = item
    
    abstract, text, informative, detailed = item  
    text_sentence_tokens = tokenize_sentences_and_words(text, remove_stop=False, lower=True)
    
    word_score_dict = word_scoring.centroid_smi(
        list(itertools.chain(*tokenize_sentences_and_words(informative, remove_stop=False, lower=True))),
        list(itertools.chain(*tokenize_sentences_and_words(detailed, remove_stop=False, lower=True))),
        lamda=0.3,
    )
    tf_score_dict = word_scoring.tf(list(itertools.chain(*text_sentence_tokens)))
    idf_score_dict = word_scoring.tf_idf(list(itertools.chain(*text_sentence_tokens)))
    smi_score_dict = word_scoring.smi(
        list(itertools.chain(*tokenize_sentences_and_words(informative, remove_stop=False, lower=True))),
        list(itertools.chain(*tokenize_sentences_and_words(detailed, remove_stop=False, lower=True))),
        lamda=-1,
        normalize=True,
        debug=False
    )
    
    sentences = tokenize_sentences_and_words(text, remove_stop=False, lower=True)
    
    word_sequence = []
    for tokens in tokenize_sentences_and_words(text, remove_stop=False, lower=True):
        for token in tokens:
            if token not in word_sequence:
                word_sequence.append(token)
                
    score_iter_callback_fn = partial(score_iter_callback_fn,
                                     word_sequence=word_sequence,
                                     abstract_tokens=list(itertools.chain(*tokenize_sentences_and_words(abstract, remove_stop=True, lower=True))),
                                     tf_score_dict=tf_score_dict,
                                     idf_score_dict=idf_score_dict,
                                     smi_score_dict=smi_score_dict,
                                    )
    
    summary = sln_summarize(
        sentences, word_score_dict, ["simplification", "diversity", "coherence"], True, make_sln_noun_verb_pronoun_prep_adj_adv, 
        score_threshold=0.3, decay_rate=0.3, desired_length=100, verbose=False, score_iter_callback_fn=score_iter_callback_fn
    )
    
#     print(summary)
    
    print(i, rouge_score(abstract, summary)["rouge-1"]["f"])
#     break
#     break
    
```

```python
abstract_set = set(itertools.chain(*tokenize_sentences_and_words(abstract, lower=True, remove_stop=False)))
text_set = set(itertools.chain(*tokenize_sentences_and_words(text, lower=True, remove_stop=False)))
all_summary = "We consider tf⋅idf score of query term in the coverage formulation. To fulfill such requirements, we define the concepts of signature term and query terms. We also can observe that the absolute improvements in overall quality and responsiveness for emails (0.9 and 0.7) is greater than for meetings (0.4 and 0.6). A node is added to the graph for each word in the sentence, and words adjacent are linked with directed edges. We connect adjacent words with directed edges. Moreover, most of the proposed systems for conversation summarization are extractive. For both datasets (meeting and email), in majority of cases (70% and 60% respectively), the users prefer the query-based abstractive summary generated by our system. Below, we show how the word graph is applied to generate the abstract sentences."

all_summary_set = set(itertools.chain(*tokenize_sentences_and_words(all_summary, lower=True, remove_stop=False)))
```

```python
both_set = abstract_set & all_summary_set
solo_abstract_set = abstract_set - both_set
solo_summary_set = all_summary_set - both_set
```

```python
def node_tag_handler(node):
    in_abstract = in_summary = False
    
    for subnode in node.split(" "):
        if subnode in both_set:
            in_abstract = in_summary = True
        elif subnode in solo_abstract_set:
            in_abstract = True
        elif subnode in solo_summary_set:
            in_summary = True
        else:
            in_summary = in_abstract = False
            break
    
    if in_summary and in_abstract:
        return "BOTH"
    if in_summary and not in_abstract:
        return "SUMMARY"
    if not in_summary and in_abstract:
        return "ABSTRACT"
    return "NODE"
```

```python
with open("text_decompose_handler.neo4j", "w") as f:
    f.write("\n".join(s.to_neo4j_expression(node_tag_handler=node_tag_handler)))
```
